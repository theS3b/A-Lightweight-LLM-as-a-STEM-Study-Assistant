```python
from datasets import Dataset, DatasetDict
import json
```

```python
# Adjust this path as necessary, it should point to the directory containing the data files from the NLP4Education repository
# (https://github.com/epfl-nlp/NLP4Education)
DATASET_LOCATION = "/home/NLP4Education/data"

with open(DATASET_LOCATION + "/courses.json") as f:
    courses = json.load(f)

with open(DATASET_LOCATION + "/questions.json") as f:
    questions = json.load(f)

# Filtering
def get_language(q):
    for c in courses:
        if q["course_id"] == c["course_id"]:
            return c["course_language"]
    raise Exception(f"Course not found: {q["course_id"]}")

LETTERS = ["A", "B", "C", "D"]

filtered_questions = []
for q in questions:
    if get_language(q).lower() == "english" and q["question_type"] == "mcq" \
    and not q["multiple_correct_answers"] and len(q["question_options"]) == 4:
        mcq_answer_index = q["mcq_answer_index"]
        if not isinstance(mcq_answer_index, list):
            mcq_answer_index = [mcq_answer_index]
        assert len(mcq_answer_index) == 1
        filtered_questions.append({
            "id": f"NLP4Education_{q["question_id"]}",
            "question": q["question_body"],
            "choices": q["question_options"],
            "answer": LETTERS[mcq_answer_index[0]]
        })

len(filtered_questions)
```
